## Q2: Continue "Q2" of the previous weeks 3 and 4

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
import pymc as pm
import arviz as az
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate synthetic normal data
np.random.seed(42)
n = 30
m, s = 0, 1
x = stats.norm(loc=m, scale=s).rvs(size=n)

# PyMC Model 1: Normal prior for theta, Gamma prior for tau
with pm.Model() as model1:
    theta = pm.Normal("theta", mu=0, sigma=1)  # Normal prior
    tau = pm.Gamma("tau", alpha=2, beta=2)     # Gamma prior
    y_obs = pm.Normal("y_obs", mu=theta, sigma=1/tau**0.5, observed=x)

    trace1 = pm.sample(2000, return_inferencedata=True, chains=2)

# PyMC Model 2: Non-normal prior for theta, Non-gamma prior for tau
with pm.Model() as model2:
    theta = pm.Cauchy("theta", alpha=0, beta=1)  # Cauchy prior
    tau = pm.Exponential("tau", lam=1)           # Exponential prior
    y_obs = pm.Normal("y_obs", mu=theta, sigma=1/tau**0.5, observed=x)

    trace2 = pm.sample(2000, return_inferencedata=True, chains=2)

# PyMC Model 3: Different non-normal & non-gamma priors
with pm.Model() as model3:
    theta = pm.StudentT("theta", nu=5, mu=0, sigma=2)  # Student-t prior
    tau = pm.HalfNormal("tau", sigma=1)               # Half-Normal prior
    y_obs = pm.Normal("y_obs", mu=theta, sigma=1/tau**0.5, observed=x)

    trace3 = pm.sample(2000, return_inferencedata=True, chains=2)

# --- DIAGNOSTICS ---
# Trace plots
az.plot_trace(trace1)
plt.suptitle("Trace Plots - Model 1 (Normal & Gamma)")
plt.show()

az.plot_trace(trace2)
plt.suptitle("Trace Plots - Model 2 (Cauchy & Exponential)")
plt.show()

az.plot_trace(trace3)
plt.suptitle("Trace Plots - Model 3 (StudentT & HalfNormal)")
plt.show()

# Posterior Predictive Checks
with model1:
    ppc1 = pm.sample_posterior_predictive(trace1)

with model2:
    ppc2 = pm.sample_posterior_predictive(trace2)

with model3:
    ppc3 = pm.sample_posterior_predictive(trace3)

# Density plots for posterior predictive checks
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

az.plot_posterior(ppc1.posterior_predictive["y_obs"], ax=axes[0], ref_val=x.mean(), hdi_prob=0.95)
axes[0].set_title("Model 1 PPC")

az.plot_posterior(ppc2.posterior_predictive["y_obs"], ax=axes[1], ref_val=x.mean(), hdi_prob=0.95)
axes[1].set_title("Model 2 PPC")

az.plot_posterior(ppc3.posterior_predictive["y_obs"], ax=axes[2], ref_val=x.mean(), hdi_prob=0.95)
axes[2].set_title("Model 3 PPC")

plt.tight_layout()
plt.show()

# Add Forest Plots for Each Model
fig, axes = plt.subplots(1, 3, figsize=(15, 4))  # Create figure but don't pass axes

az.plot_forest(trace1, var_names=["theta", "tau"], combined=False, hdi_prob=0.95, r_hat=True)
plt.title("Model 1 Forest Plot")

az.plot_forest(trace2, var_names=["theta", "tau"], combined=False, hdi_prob=0.95, r_hat=True)
plt.title("Model 2 Forest Plot")

az.plot_forest(trace3, var_names=["theta", "tau"], combined=False, hdi_prob=0.95, r_hat=True)
plt.title("Model 3 Forest Plot")

plt.tight_layout()
plt.show()


```

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

# Slice Sampling and Its Role in Metropolis-Within-Gibbs

## Understanding Slice Sampling

Slice sampling is an efficient Markov Chain Monte Carlo (MCMC) method that generates samples from a target distribution without requiring an explicit proposal distribution, unlike the Metropolis-Hastings algorithm. The core idea behind slice sampling is to introduce an auxiliary variable and then sample iteratively to obtain values from the desired distribution.

The method works as follows:

1. **Starting Point**: Begin with a given value \\( x_t \\) from the current iteration of the Markov chain.
2. **Introduce an Auxiliary Variable**: Draw \\( y \sim U(0, f(x_t)) \\). This selects a random horizontal slice beneath the probability density curve of \\( f(x) \\).
3. **Determine the Slice Interval**: Identify an interval \\( [a, b] \\) that contains \\( x_t \\) where \\( f(x) > y \\). This means that only values within this interval are valid candidates for the next sample.
4. **Sample from the Interval**: Draw a new candidate \\( x' \\) uniformly from \\( [a, b] \\).
5. **Acceptance and Iteration**: If \\( f(x') > y \\), accept \\( x' \\) as the new sample. If not, reduce the interval and repeat the process until an acceptable sample is found.

This process allows slice sampling to adaptively explore the distribution, making it particularly useful when the full normalizing constant of the target distribution is unknown or difficult to compute.

## Explanation of the `slice_f_at_y` Function

The provided Python function `slice_f_at_y(f, x, y)` implements the slice sampling step. Heres how it works:

- First, it identifies the interval \\( [a, b] \\) where \\( f(x) > y \\). The function searches for the leftmost and rightmost values within a predefined grid where the function evaluates above \\( y \\).
- Next, it samples a new \\( x' \\) uniformly from the identified interval.
- If the new sample \\( x' \\) satisfies \\( f(x') > y \\), it is accepted immediately.
- If not, the function narrows the interval and recursively tries again using `slice_f_at_y_`, which refines the sampling region iteratively until an acceptable value is found.

This method avoids the need for explicit tuning parameters, as it dynamically adjusts the interval width based on the function's behavior.

## Using Slice Sampling in Metropolis-Within-Gibbs

Metropolis-within-Gibbs is a hybrid MCMC method that combines Gibbs sampling with Metropolis-Hastings steps. In scenarios where full conditional distributions are difficult to sample directly, Metropolis-Hastings is often used. However, an alternative approach is to use slice sampling in place of Metropolis-Hastings, providing a more automated and often more efficient solution.

To replace Metropolis-Hastings with slice sampling in a Metropolis-within-Gibbs framework, the following steps are taken:

1. **Define the Full Conditional Distribution**: The parameter being sampled follows a full conditional distribution known only up to a normalizing constant.
2. **Set Up Slice Sampling**:
   - Generate an auxiliary variable \\( y \sim U(0, f(x)) \\), selecting a height under the density curve.
   - Find the interval \\( [a, b] \\) where \\( f(x) > y \\), defining the valid slice.
   - Sample a new value \\( x' \\) uniformly from this interval.
3. **Acceptance Without Rejection**: Unlike Metropolis-Hastings, which requires an acceptance step, slice sampling ensures that \\( x' \\) is always drawn from the valid region, eliminating the need for explicit rejection.
4. **Iterate Over Parameters**: The process is repeated for each parameter in the Gibbs sampling cycle, ensuring that all variables are updated in turn.

## The Curve Under Which We Sample

The key concept in slice sampling is that we sample from a region beneath the probability density function (PDF) of the full conditional distribution. The auxiliary variable \\( y \\) defines a horizontal slice of this density, and the sampling step ensures that the next draw remains within this slice. Since the exact normalization of the conditional distribution is not needed, slice sampling is particularly useful in hierarchical Bayesian models where the full conditionals are known only up to a constant.

## Initial Values and Steps for Gibbs Sampling

When incorporating slice sampling into Gibbs sampling, the initial values must be chosen appropriately, often based on prior information or reasonable defaults. Each iteration consists of:

1. Selecting a variable to update while keeping all others fixed.
2. Drawing a new sample using slice sampling, ensuring the sample is within the valid density region.
3. Updating the Markov chain with the newly sampled value.
4. Moving to the next parameter and repeating the process.

This iterative process continues until convergence, producing samples from the joint posterior distribution.

## Advantages of Slice Sampling Over Metropolis-Hastings

1. **No Need for a Proposal Distribution**: Unlike Metropolis-Hastings, which requires careful tuning of the proposal distribution, slice sampling automatically adjusts based on the target functions shape.
2. **Handles Unnormalized Densities**: Since slice sampling only requires function evaluations relative to an unknown constant, it is well-suited for Bayesian posterior distributions.
3. **Adaptive Exploration**: The method dynamically determines the step size, avoiding issues related to poorly chosen step lengths in Metropolis-Hastings.
4. **Guaranteed Acceptance**: Once a valid sample is found within the slice, it is always accepted, reducing wasted computations.

## Conclusion

Slice sampling is a powerful alternative to Metropolis-Hastings in Metropolis-within-Gibbs algorithms, especially when full conditionals are only known up to a normalizing constant. By efficiently exploring the target distribution and avoiding tuning difficulties, slice sampling provides a robust and automated approach to Bayesian inference. Implementing this method can improve sampling efficiency and reliability in complex hierarchical models.
